```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define the momentum update function
def momentum_update(grads, beta=0.9, learning_rate=0.01, initial_value=0.0):
    momentum = 0.0
    value = initial_value
    values = [value]
    
    for grad in grads:
        momentum = beta * momentum + (1 - beta) * grad
        value -= learning_rate * momentum
        values.append(value)
    
    return values

# Create two lists of gradient updates
# consistent_grads = [0.1] * 10
# inconsistent_grads = [0.1, -0.1] * 5

consistent_grads = np.random.uniform(0, 3, 50)
inconsistent_grads = np.random.randn(50) 

# Perform the momentum updates
consistent_values = momentum_update(consistent_grads)
inconsistent_values = momentum_update(inconsistent_grads)

```

```python

# Define a regular Gradient Descent update function
def reg_gd_updates(grads, learning_rate = 0.01, initial_value = 0.0):

    value = initial_value
    values = [value]

    for grad in grads:
        value -= learning_rate * grad
        values.append(value)

    return values


regular_GD_values_cv = reg_gd_updates(consistent_grads)
regular_GD_values_iv = reg_gd_updates(inconsistent_grads)
```

```python
# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(consistent_values, marker='o')
plt.title('Consistent Gradient Updates')
plt.xlabel('Iteration')
plt.grid()
plt.ylabel('Parameter Value')

plt.subplot(2, 2, 2)
plt.plot(inconsistent_values, marker='o')
plt.title('Inconsistent Gradient Updates')
plt.xlabel('Iteration')
plt.grid()
plt.ylabel('Parameter Value')

plt.subplot(2, 2, 3)
plt.plot(regular_GD_values_cv, marker='o')
plt.title('GD consistent Gradient Updates')
plt.xlabel('Iteration')
plt.grid()
plt.ylabel('Parameter Value')

plt.subplot(2, 2, 4)
plt.plot(regular_GD_values_iv, marker='o')
plt.title('GD Inconsistent Gradient Updates')
plt.xlabel('Iteration')
plt.grid()
plt.ylabel('Parameter Value')

plt.tight_layout()
plt.show()
```
